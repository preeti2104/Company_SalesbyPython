# EXPLORATORY DATA ANALYSIS

# Gender

```python
df.columns
```

```python
colors=['pink','red']
ax = sns.countplot(x ='Gender', data =df,palette = colors) #to get the chart
for bars in ax.containers: # to get the values
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sales_gen=df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x= 'Gender',y ='Amount',data = sales_gen,palette=colors)
```

From the above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than the men

# Age

```python
colors=['pink','salmon']
ax =sns.countplot(data = df,x='Age Group',hue='Gender',palette=colors)
```

```python
ax = sns.countplot(data = df,x ='Age Group',hue='Gender',palette=colors)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index= False)['Amount'].sum().sort_values(by ='Amount',ascending = False)
sns.barplot(x='Age Group',y='Amount',data = sales_age,palette=colors)
```

From the above graphs we can see that most of the buyers are of the age group between 26-35 years female

# State

```python
colors=['pink','salmon','coral','tomato','crimson']
sales_state= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by ='Orders',ascending = False).head(10)
sns.set(rc = {'figure.figsize':(16,6)})
sns.barplot(data = sales_state,x='State',y = 'Orders',palette=colors)
```

```python
colors=['pink','salmon','coral','tomato','crimson']
sales_state= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending = False).head(10)
sns.set(rc = {'figure.figsize':(16,6)})
sns.barplot(data = sales_state,x='State',y = 'Amount',palette=colors)
```

From the above graphs we can see that most of the orders and total sales/amunt are from Uttar Pradesh,Maharashtra and Karnataka

# Marital Status 

```python
colors=['pink','salmon','coral','tomato','crimson']
ax = sns.countplot(data = df, x='Marital_Status',palette=colors)
sns.set(rc = {'figure.figsize':(2,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_data,x='Marital_Status',y = 'Amount',hue = 'Gender')
```

From the above graphs we can see that most of the buyers are married(women) and they have high purchasing power

# Occupation

```python
colors=['pink','salmon','coral','tomato','crimson']
ax = sns.countplot(data = df, x='Occupation',palette=colors)
sns.set(rc = {'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = sales_data,x='Occupation',y = 'Amount')
```

From the graphs we can see that most of the buyers are working in IT,Healthcare and Aviation

# Product Category

```python
colors=['tomato','crimson','pink','salmon','coral']
sns.set(rc={'figure.figsize':(25,5)})
ax= sns.countplot(data = df,x='Product_Category',palette=colors)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
colors=['tomato','crimson','pink','salmon','coral']
sales_data = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc ={'figure.figsize':(25,5)})
sns.barplot(data=sales_data,x='Product_Category',y='Amount',palette=colors)
```

From the above graphs we can see that most of the sold products are from Food,Clothing and Electronics category
